### Text Pre-Processing and Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Text Processing done in tfidf vectorizer itself
# max_features=50000 --> keeps only the 50000 most used tokens
# stop_words='english' --> removes all stop words from corpus
# lowercase=True --> converts all text to lowercase
# ngram_range=(1,2) --> utilizes bi-grams. Two words together rather than singular words
# token_pattern=r'(?u)\b[a-zA-Z]{2,}\b' --> Removes punctuation automatically, Keeps words with length ≥ 2, Removes most special characters implicitly
# min_df=3 --> removes words/tokens that do not occur in >=3 records
tfidf = TfidfVectorizer(
    max_features=50000,
    stop_words='english',
    lowercase=True,
    ngram_range=(1,2),
    token_pattern=r'(?u)\b[a-zA-Z]{2,}\b',
    min_df=3
)

vectors = tfidf.fit_transform(df['tags'])
```

### Recommendation Model

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
'''pickle.dump(similarity, open('similarity.pkl', 'wb'))'''
```

```python
similarity[0][1]  # similarity between movie 0 and movie 1
```

```python
def recommend(movie_title):
    if movie_title not in df['title'].values:
        return "Movie not found"
    
    idx = df[df['title'] == movie_title].index[0] #Returns index value of the movie entered in
    
    sim_scores = cosine_similarity(vectors[idx].reshape(1, -1), vectors)[0] # Finds similarity score of all values in our vectors wrt movie_title index
    
    sim_scores = list(enumerate(sim_scores)) # store similarity scores along with index
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:11] # Store only the top 10 movies with lowest similarity score/distance. Exclude the first as this will coincide with the movie_title itself
    
    movie_indices = [i[0] for i in sim_scores] # return the index values of sim_scores
    
    return df['title'].iloc[movie_indices] # Filter titles from dataframe
```

### Testing Recoomendation on a sample of Movies
