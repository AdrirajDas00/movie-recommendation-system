```python
import pandas as pd
import pickle
```

```python
df = pickle.load(open('movies_with_posters.pkl', 'rb'))
```

```python
# Retreive poster URL from OMDB API.
# Free version allows 1000 API calls a day, therefore needs to be done in batches. 5000 movies, 5000/1000=5days 
import requests
from tqdm import tqdm
for i, row in df.iterrows():
    if pd.isna(df.loc[i, 'poster_url']):  # skip already done
        try:
            data = requests.get(row['omdb_link']).json()
            df.loc[i, 'poster_url'] = data.get('Poster', None)
        except:
            df.loc[i, 'poster_url'] = None

        # Save every 50 rows
        if i % 50 == 0:
            df.to_pickle('movies_with_posters.pkl')
```

```python
# Save into movies.pkl
pickle.dump(df, open('movies.pkl', 'wb'))
```
