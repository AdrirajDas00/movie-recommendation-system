```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Reading Metadata File
# Dataset Link: https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset
df=pd.read_csv('../movies_metadata.csv')
df.head()
```

```python
df.dtypes
```

```python
# id column unclean. keeping only the records that are clean (ID is numeric)
# Also Filtering out records with imdb_id empty (Required for attaining Movie Posters)
df['id'] = pd.to_numeric(df['id'], errors='coerce')
df = df.dropna(subset=['id'])
df['id'] = df['id'].astype('int32')
df = df[df['imdb_id'].notnull()]
```

```python
df.id.dtypes
```

```python
# sorting movies based on vote count (popularity)
df=df.sort_values(by='vote_count',ascending=False)
df.head()
```

```python
# Filtering out only the top 5000 popular movies
df_5000=df.head(5000)
```

```python
# Filtering out only the relevant columns
df_5000_filteredcols=df_5000['id,title,imdb_id,overview,genres,tagline,production_companies'.split(',')]
df_5000_filteredcols.head()
```

```python
# Reading credits file
df_credits=pd.read_csv('../credits.csv')
df_credits.head()
```

```python
# cleaning the id column (keeping records where id is nemeric)
df_credits['id'] = pd.to_numeric(df_credits['id'], errors='coerce')
df_credits = df_credits.dropna(subset=['id'])
df_credits['id'] = df_credits['id'].astype('int32')
```

```python
# Performing inner join with df_5000_filteredcols dataframe. Should take into consideration the top 5000 movies previously filtered out from df_credits
df_combined=pd.merge(df_5000_filteredcols,df_credits,on='id',how='inner')
df_combined
```

```python
df_clean=df_combined.copy()
```

```python
# Performing JSON parsing with ast. Columns genres and production companies have data stored within the 'name' field.
import ast
def extract_names(obj):
    try:
        return ", ".join([i['name'] for i in ast.literal_eval(obj)]) # Seperate values with a comma
    except:
        return ""

df_clean['genres'] = df_clean['genres'].apply(extract_names)
df_clean['production_companies'] = df_clean['production_companies'].apply(extract_names)
```

```python
df_clean
```

```python
# Performing JSON parsing for cast column. Since, cast list can be massive, only the first 5 cast members (main cast) will be taken into consideration
def get_top_cast(cast_str):
    try:
        cast_list = ast.literal_eval(cast_str)
        names = [member['name'] for member in cast_list[:5]]
        return ", ".join(names)
    except:
        return ""

# Performing JSON parsing for director column
def get_director(crew_str):
    try:
        crew_list = ast.literal_eval(crew_str)
        for member in crew_list:
            if member['job'] == 'Director':
                return member['name']
        return ""
    except:
        return ""
```

```python
df_clean['cast'] = df_clean['cast'].apply(get_top_cast)
df_clean['director'] = df_clean['crew'].apply(get_director)
```

```python
# dropping crew column
df_clean.drop(columns='crew',inplace=True)
```

```python
# creating tags column which takes into consideration overview, genres, cast, director and tagline columns. This column will be our key data field for text preparation and model building
df_clean['tags'] = (
    df_clean['overview'] + ' ' +
    df_clean['genres'] + ' ' +
    df_clean['cast'] + ' ' +
    df_clean['director'] + ' ' +
    df_clean['tagline']
)
```
